```python
import pickle 
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("diabetes_model.pkl","rb") as file:
    loaded_model=pickle.load(file)
print("model loaded successfully")    
```

```python
print("Enter the patient details below:")
Pregnancies = float(input("Number of Pregnancies: "))
Glucose = float(input("Glucose Level: "))
BloodPressure = float(input("Blood Pressure: "))
SkinThickness = float(input("Skin Thickness: "))
Insulin = float(input("Insulin Level: "))
BMI = float(input("BMI: "))
DiabetesPedigreeFunction = float(input("Diabetes Pedigree Function: "))
Age = float(input("Age: "))
```

```python
input_data = np.array([[Pregnancies, Glucose, BloodPressure, SkinThickness, Insulin, BMI, DiabetesPedigreeFunction, Age]])
```

```python
prediction = loaded_model.predict(input_data)
```

```python
if prediction[0] == 1:
    print("\n The model predicts that the patient is likely to have Diabetes.")
else:
    print("\n The model predicts that the patient is NOT likely to have Diabetes.")
```

```python
if prediction[0] == 1:
    color = 'red'
    text = "Diabetes"
else:
    color = 'green'
    text = "Not Diabetes"


fig, ax = plt.subplots()
circle = plt.Circle((0.5, 0.5), 0.3, color=color)
ax.add_patch(circle)
ax.text(0.5, 0.5, text, fontsize=14, color='white', ha='center', va='center', fontweight='bold')

# Adjust axes
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal')
ax.axis('off')

plt.show()
```
